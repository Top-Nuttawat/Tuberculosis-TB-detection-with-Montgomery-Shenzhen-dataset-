# Montgomery dataset

```python

download_folder = "Montgomery_Set/ClinicalReadings/"
os.makedirs(download_folder, exist_ok=True)

Montgomery_url = "https://data.lhncbc.nlm.nih.gov/public/Tuberculosis-Chest-X-ray-Datasets/Montgomery-County-CXR-Set/MontgomerySet/ClinicalReadings/"

# file_list = []
# for i in range(1, 401): # Montgomery dataset max MCUCXR_0399_1.txt
#     for suffix in ["0", "1"]:
#         file_list.append(f"MCUCXR_{i:04d}_{suffix}.txt")
        
# Montgomery dataset  max MCUCXR_0399_1.txt
file_list = [f"MCUCXR_{i:04d}_{suffix}.txt" for i in range(1, 401) for suffix in ["0", "1"]]

for filename in tqdm(file_list, desc="Downloading files"):
    file_url = Montgomery_url + filename
    r = requests.get(file_url)
    if r.status_code == 200:
        with open(os.path.join(download_folder, filename), "wb") as f:
            f.write(r.content)
```

```python
# URL montgomery_consensus_roi.csv
csv_url = "https://data.lhncbc.nlm.nih.gov/public/Tuberculosis-Chest-X-ray-Datasets/Montgomery-County-CXR-Set/MontgomerySet/montgomery_consensus_roi.csv"

filename = os.path.basename(csv_url)

# Created folder
download_folder = "Montgomery_Set"
os.makedirs(download_folder, exist_ok=True)

# Download montgomery_consensus_roi.csv
response = requests.get(csv_url)
with open(os.path.join(download_folder, filename), "wb") as f:
    f.write(response.content)

Montgomery_df = pd.read_csv('/kaggle/working/Montgomery_Set/montgomery_consensus_roi.csv')
Montgomery_df.head()
```

# 1. EDA (Exploratory Data Analysis)

```python
mont_image_list = os.listdir('/kaggle/working/Montgomery_Set/CXR_png')
shen_image_list = os.listdir('/kaggle/working/Shenzhen_Set/CXR_png')
```

```python
print(len(mont_image_list))
print(len(shen_image_list))
```

```python
# put the images into dataframes
df_mont = pd.DataFrame(mont_image_list, columns=['image_id'])
df_shen = pd.DataFrame(shen_image_list, columns=['image_id'])

# remove the 'Thunbs.db' line
df_mont = df_mont[df_mont['image_id'] != 'Thumbs.db']
df_shen = df_shen[df_shen['image_id'] != 'Thumbs.db']

# Reset the index or this will cause an error later

df_mont.reset_index(inplace=True, drop=True)
df_shen.reset_index(inplace=True, drop=True)

print(df_mont.shape)
print(df_shen.shape)
```
